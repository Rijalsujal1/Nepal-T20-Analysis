 Comprehensive Match Outcome Analysis

  This code block transforms the raw, ball-by-ball data into a structured, match-by-match summary to analyze Nepal's overall T20I performance. The process is broken down into several key steps:

   1. Aggregate Innings Data: The code first filters out non-standard innings (like super overs) and then groups the data by each match and innings. This allows for the calculation of the total
      runs scored and wickets lost for each team in their respective innings.


   2. Create a Match-Level View: Using pivot_table, the data is reshaped so that each row represents a single match. The columns are flattened and renamed to clearly show the scores, wickets,
      and teams for both the first and second innings (e.g., innings1_score, innings2_batting_team).


   3. Determine Match Winners: To efficiently determine the outcome of every match, the code uses np.select. This vectorized operation compares the scores from both innings to populate winner,
      loser, and match_outcome columns for the entire dataset, handling wins, ties, and no-result matches.

   4. Filter for Nepal's Matches: A new DataFrame, nepal_matches, is created by filtering the main match list to include only those games where Nepal was either the batting or bowling team in
      either innings.


   5. Calculate Nepal's Final Results: A function is applied to the nepal_matches DataFrame to determine Nepal's specific result in each match (Win, Loss, Tie, or No Result). This is then used
      to generate the final summary, which counts the total number of wins, losses, ties, and no-result matches for Nepal across the entire dataset.

```python
import numpy as np

# STEP 1: Filter standard innings (1 and 2)
standard_df = df[df['innings'].isin([1, 2])].copy()

# STEP 2: Aggregate total runs and wickets per innings
innings_summary = standard_df.groupby(
    ['match_no', 'innings', 'batting_team', 'bowling_team']
).agg(
    total_runs=('total_runs', 'sum'),
    wickets_lost=('player_out', lambda x: x.notna().sum())
).reset_index()

# STEP 3: Pivot to separate innings into columns
match_pivot = innings_summary.pivot_table(
    index='match_no',
    columns='innings',
    values=['total_runs', 'wickets_lost', 'batting_team', 'bowling_team'],
    aggfunc='first'
).reset_index()

# Flatten multi-level columns
match_pivot.columns = ['_'.join(filter(None, map(str, col))) for col in match_pivot.columns]

# STEP 4: Rename columns for clarity
match_pivot.rename(columns={
    'total_runs_1': 'innings1_score',
    'total_runs_2': 'innings2_score',
    'wickets_lost_1': 'innings1_wickets',
    'wickets_lost_2': 'innings2_wickets',
    'batting_team_1': 'innings1_batting_team',
    'bowling_team_1': 'innings1_bowling_team',
    'batting_team_2': 'innings2_batting_team',
    'bowling_team_2': 'innings2_bowling_team'
}, inplace=True)

# STEP 5: Determine match outcome using np.select
conditions = [
    match_pivot['innings2_score'].isna(),
    match_pivot['innings1_score'] == match_pivot['innings2_score'],
    match_pivot['innings2_score'] > match_pivot['innings1_score'],
]
winner_outcomes = [None, None, match_pivot['innings2_batting_team']]
loser_outcomes = [None, None, match_pivot['innings1_batting_team']]

match_pivot['winner'] = np.select(conditions, winner_outcomes, default=match_pivot['innings1_batting_team'])
match_pivot['loser'] = np.select(conditions, loser_outcomes, default=match_pivot['innings2_batting_team'])
match_pivot.loc[match_pivot['innings2_score'].isna(), ['winner', 'loser']] = None
match_pivot.loc[match_pivot['innings1_score'] == match_pivot['innings2_score'], ['winner', 'loser']] = None

# STEP 6: Filter matches involving Nepal
nepal_matches = match_pivot[
    (match_pivot['innings1_batting_team'] == 'Nepal') |
    (match_pivot['innings1_bowling_team'] == 'Nepal') |
    (match_pivot['innings2_batting_team'] == 'Nepal') |
    (match_pivot['innings2_bowling_team'] == 'Nepal')
].copy()

# STEP 7: Determine Nepal's result per match using a fully vectorized method
nepal_conditions = [
    nepal_matches['winner'] == 'Nepal',
    nepal_matches['loser'] == 'Nepal',
    (nepal_matches['winner'].isna()) & (nepal_matches['loser'].isna()) & (nepal_matches['innings1_score'] == nepal_matches['innings2_score']),
    nepal_matches['innings2_score'].isna()
]
nepal_outcomes = ['Win', 'Loss', 'Tie', 'No Result']
nepal_matches['nepal_result'] = np.select(nepal_conditions, nepal_outcomes, default='Unknown')

# Summary of Nepal's match results
nepal_results_summary = nepal_matches['nepal_result'].value_counts()

print("\nNepal's Match Result Summary:")
print(nepal_results_summary)
```

```python
nepal_bowling_df = df[df['bowling_team'] == 'Nepal'].copy()
wickets_df = nepal_bowling_df.dropna(subset=['player_out'])
wickets_df = wickets_df[wickets_df['dismissal_kind'] != 'run out'].copy()

total_wickets_per_bowler = wickets_df.groupby('bowler').size()
qualified_bowlers = total_wickets_per_bowler[total_wickets_per_bowler >= 10]

sorted_bowlers = qualified_bowlers.sort_values(ascending=False).index

qualified_wickets_df = wickets_df[wickets_df['bowler'].isin(sorted_bowlers)].copy()

def get_phase(over):
    if over in range(0, 6):
        return 'Powerplay (1-6)'
    elif over in range(6, 15):
        return 'Middle Overs (7-15)'
    elif over in range(15, 20):
        return 'Death Overs (16-20)'
    return 'Other'

qualified_wickets_df['phase'] = qualified_wickets_df['over'].apply(get_phase)

phase_pivot = pd.crosstab(qualified_wickets_df['bowler'], qualified_wickets_df['phase'])
phase_pivot = phase_pivot.reindex(index=sorted_bowlers)

phase_order = ['Powerplay (1-6)', 'Middle Overs (7-15)', 'Death Overs (16-20)']
phase_pivot = phase_pivot.reindex(columns=phase_order, fill_value=0)

plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(12, 10))

ax = sns.heatmap(
    phase_pivot,
    annot=True,
    fmt='d',
    cmap='YlGnBu',
    linewidths=.5,
    cbar_kws={'label': 'Number of Wickets'}
)

ax.set_title(
    "Phase-wise Wicket Distribution for Nepali Bowlers",
    fontsize=18,
    fontweight='bold',
    pad=20
)
ax.text(ax.get_xlim()[1], ax.get_ylim()[0] - 0.5, '(Min. 10 Total Wickets)', ha='right', fontsize=12, style='italic')

ax.set_xlabel("Innings Phase", fontsize=14)
ax.set_ylabel("Bowler", fontsize=14)

plt.xticks(fontsize=12)
plt.yticks(rotation=0, fontsize=12)

plt.tight_layout()
plt.show()
```

```python
nepal_batting_df = df[df['batting_team'] == 'Nepal'].copy()

def get_phase(over):
    if over in range(0, 6): return 'Powerplay (1-6)'
    elif over in range(6, 15): return 'Middle Overs (7-15)'
    elif over in range(15, 20): return 'Death Overs (16-20)'
    return 'Other'

nepal_batting_df['phase'] = nepal_batting_df['over'].apply(get_phase)
nepal_batting_df = nepal_batting_df[nepal_batting_df['phase'] != 'Other']

total_balls_faced = nepal_batting_df[nepal_batting_df['wides'] == 0].groupby('batsman').size()
qualified_batsmen_list = total_balls_faced[total_balls_faced >= 100].index
qualified_df = nepal_batting_df[nepal_batting_df['batsman'].isin(qualified_batsmen_list)]

phase_stats = qualified_df.groupby(['batsman', 'phase']).agg(
    runs=('batsman_runs', 'sum'),
    balls=('wides', lambda x: (x == 0).sum()),
    dismissals=('player_out', lambda x: x.notna().sum())
).reset_index()

phase_stats['average'] = phase_stats.apply(
    lambda row: row['runs'] / row['dismissals'] if row['dismissals'] > 0 else None,
    axis=1
)
phase_stats['strike_rate'] = phase_stats.apply(
    lambda row: (row['runs'] / row['balls']) * 100 if row['balls'] > 0 else 0,
    axis=1
)
phase_stats.dropna(subset=['average'], inplace=True)

plt.style.use('seaborn-v0_8-whitegrid')
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 9))
fig.suptitle("Phase-wise Batting Performance for Nepali Players", fontsize=22, fontweight='bold')

sns.stripplot(
    data=phase_stats,
    x='phase',
    y='average',
    ax=ax1,
    palette='viridis',
    jitter=0.2,
    size=7,
    alpha=0.7,
    order=['Powerplay (1-6)', 'Middle Overs (7-15)', 'Death Overs (16-20)']
)
ax1.set_title('Batting Average Distribution by Phase', fontsize=16, pad=15)
ax1.set_xlabel('Innings Phase', fontsize=14)
ax1.set_ylabel('Batting Average (Runs per Dismissal)', fontsize=14)
ax1.tick_params(axis='x', labelsize=12)
ax1.tick_params(axis='y', labelsize=12)

sns.stripplot(
    data=phase_stats,
    x='phase',
    y='strike_rate',
    ax=ax2,
    palette='plasma',
    jitter=0.2,
    size=7,
    alpha=0.7,
    order=['Powerplay (1-6)', 'Middle Overs (7-15)', 'Death Overs (16-20)']
)
ax2.set_title('Strike Rate Distribution by Phase', fontsize=16, pad=15)
ax2.set_xlabel('Innings Phase', fontsize=14)
ax2.set_ylabel('Strike Rate (Runs per 100 Balls)', fontsize=14)
ax2.tick_params(axis='x', labelsize=12)
ax2.tick_params(axis='y', labelsize=12)

sns.despine()
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python

nepal_batting_df = df[df['batting_team'] == 'Nepal'].copy()
batting_appearances = nepal_batting_df[['match_no', 'batsman']].rename(columns={'batsman': 'player'})
non_striker_appearances = nepal_batting_df[['match_no', 'non_striker']].rename(columns={'non_striker': 'player'})


nepal_bowling_df = df[df['bowling_team'] == 'Nepal'].copy()
bowling_appearances = nepal_bowling_df[['match_no', 'bowler']].rename(columns={'bowler': 'player'})

all_nepali_appearances = pd.concat([
    batting_appearances,
    non_striker_appearances,
    bowling_appearances
]).dropna(subset=['player'])

matches_played = (
    all_nepali_appearances
    .drop_duplicates()
    .groupby('player')['match_no']
    .nunique()
    .reset_index()
)
matches_played.columns = ['player', 'matches_played']



standard_catches = (
    nepal_bowling_df[nepal_bowling_df['dismissal_kind'] == 'caught']
    .dropna(subset=['fielders'])['fielders']
    .value_counts()
)

c_and_b_catches = (
    nepal_bowling_df[nepal_bowling_df['dismissal_kind'] == 'caught and bowled']['bowler']
    .value_counts()
)

total_catches = standard_catches.add(c_and_b_catches, fill_value=0).astype(int).reset_index()
total_catches.columns = ['player', 'total_catches']

# --- Step 3: Combine Data and Prepare for Plotting ---

fielding_stats_df = pd.merge(
    matches_played,
    total_catches,
    on='player',
    how='left'
)
fielding_stats_df['total_catches'] = fielding_stats_df['total_catches'].fillna(0).astype(int)

# Identify the top 10 players by catches
top_10_by_catches = fielding_stats_df.sort_values(by='total_catches', ascending=False).head(10)

# Melt the DataFrame for grouped bar plotting
plot_df = top_10_by_catches.melt(
    id_vars='player',
    value_vars=['matches_played', 'total_catches'],
    var_name='Metric',
    value_name='Count'
)

# --- Step 4: Create the Grouped Bar Plot ---

plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(15, 8))

sns.barplot(
    data=plot_df,
    x='player',
    y='Count',
    hue='Metric',
    palette={'matches_played': 'skyblue', 'total_catches': 'coral'},
    ax=ax
)

# Add text labels to each bar
for container in ax.containers:
    ax.bar_label(container, fontsize=10)

# --- Final Styling ---

ax.set_title(
    "Catches vs. Matches Played for Top 10 Nepali Fielders",
    fontsize=20,
    fontweight='bold',
    pad=20
)
ax.set_xlabel("Player", fontsize=14)
ax.set_ylabel("Total Count", fontsize=14)
ax.legend(title='Metric')

# Rotate x-axis labels for better readability
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

sns.despine()
plt.tight_layout()

plt.savefig("catches_vs_matches_barplot_final.png", dpi=300, bbox_inches='tight')
plt.show()
```
